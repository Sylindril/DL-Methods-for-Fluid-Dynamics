```python
import numpy as np
import matplotlib.pyplot as plt

fft2 = np.fft.fft2
ifft2 = np.fft.ifft2
```

```python
N = 100
L=N

kxx = 2*np.pi*np.fft.fftfreq(N)
kx, ky = np.meshgrid(kxx, kxx)
```

```python
x, y = np.meshgrid(np.linspace(0, L, N), np.linspace(0, L, N))
rr = np.sqrt( ((x-L/2)*(x-L/2)+(y-L/2)*(y-L/2))*.5 )

sig = (np.exp(-0.1*rr))

plt.pcolor(x, y, sig, cmap=plt.cm.Blues)
```

```python
sig_k=fft2(sig)
```

```python
xD, yD = 30, 0
kdotr = kx*xD + ky*yD
```

```python
sig_new = sig_k*np.exp(-1j*kdotr)
```

```python
plt.pcolor(x, y, np.real(ifft2(sig_new)), cmap=plt.cm.Blues)
```
